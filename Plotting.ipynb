```python
import pandas as pd
import matplotlib.pyplot as plt

recall_df = pd.read_csv("results/recall.txt").set_index("neighbors")
precision_df = pd.read_csv("results/precision.txt").set_index("neighbors")
f1_df = pd.read_csv("results/f1.txt").set_index("neighbors")
```

```python
recall_df.plot(title = "(Iris) Recall Over Neighbors")

plt.plot
plt.ylabel("Recall")
plt.xlabel("Neighbors")
```

```python
precision_df.plot(title = "(Iris) Precision Over Neighbors")
plt.plot
plt.ylabel("Precision")
plt.xlabel("Neighbors")
```

```python
f1_df.plot(title = "(Iris) F1 Score Over Neighbors")
plt.plot
plt.ylabel("F1 Score")
plt.xlabel("Neighbors")
```
